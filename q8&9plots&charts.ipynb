```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
x=[0,1,2,3,4,5]
y=[0,1,4,9,16,25]
plt.plot(x,y)
plt.title("Positive part of parabola")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.show()
```

```python
plt.scatter(x,y,color="red", marker="*")
plt.legend()
plt.xlabel("x")
plt.ylabel("f(x)")
plt.show()
```

```python
plt.scatter(x,y,color="red", marker="*")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.title("stock price prediction")
plt.show()
```

```python
plt.bar(x,y,0.25,color="red",label="x")
plt.show()
```

```python
labels=["cloud computing","big data","IOT","ML-AI"]
values=[5,65,25,15]
plt.pie(values,labels=labels, radius=1)
plt.show()
plt.pie(values,labels=labels, radius=1, explode=(0,0,1,0),shadow=True)
plt.show()
```

```python
x=np.arange(0,4*np.pi,0.1)
y=np.sin(x)
z=np.cos(x)
plt.plot(x,y,x,z)
plt.xlabel('x values from 0 to 4pi')
plt.ylabel('sin(x) and cos(x)')
plt.title('Plot of sin and cos from 0 to 4pi')
plt.legend(['sin(x)','cos(x)'])
plt.show()
```

```python
x=np.arange(0,8*np.pi,0.1)
k=np.tan(x)
plt.plot(x,k)
plt.xlabel('x values from 0 to 8pi')
plt.ylabel('Tan(x)')
plt.title('Plot of tan from 0 to 8pi')
plt.legend(['tan(x)'])
plt.show()
```
